## Query 7: INNER JOIN

Join two tables to combine category and department information.

```python
query = """
SELECT
    s.Category,
    c.Department,
    s.Sales
FROM sales_data s
INNER JOIN category_info c
ON s.Category = c.Category
LIMIT 10;
"""

pd.read_sql(query, engine)
```

## Query 8: Subquery

Retrieve records having sales greater than the average sales.

```python
query = """
SELECT *
FROM sales_data
WHERE Sales >
(
    SELECT AVG(Sales)
    FROM sales_data
);
"""

pd.read_sql(query, engine)
```

## Query 9: Common Table Expression (CTE)

Calculate total sales for each category using the WITH clause.

```python
query = """
WITH CategorySales AS
(
    SELECT
        Category,
        SUM(Sales) AS TotalSales
    FROM sales_data
    GROUP BY Category
)

SELECT *
FROM CategorySales;
"""

pd.read_sql(query, engine)
```

## Query 10: ROW_NUMBER()

Assign row numbers to sales records within each category.

```python
query = """
SELECT
Category,
Sales,
ROW_NUMBER() OVER
(
PARTITION BY Category
ORDER BY Sales DESC
) AS Row_Number
FROM sales_data;
"""

pd.read_sql(query, engine)
```

## Query 11: RANK()

Rank sales within each category.

```python
query = """
SELECT
Category,
Sales,
RANK() OVER
(
PARTITION BY Category
ORDER BY Sales DESC
) AS Sales_Rank
FROM sales_data;
"""

pd.read_sql(query, engine)
```

## Query 12: LAG()

Display previous sales value.

```python
query = """
SELECT
"Order Date",
Sales,
LAG(Sales) OVER
(
ORDER BY "Order Date"
) AS Previous_Sales
FROM sales_data;
"""

pd.read_sql(query, engine)
```

## Query 13: LEAD()

Display next sales value.

```python
query = """
SELECT
"Order Date",
Sales,
LEAD(Sales) OVER
(
ORDER BY "Order Date"
) AS Next_Sales
FROM sales_data;
"""

pd.read_sql(query, engine)
```
