## Step 1: Import Required Libraries

Import the necessary libraries for data analysis and visualization.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import os
```

## Step 2: Load the Cleaned Dataset

Load the cleaned Superstore Sales dataset into a Pandas DataFrame.

```python
df = pd.read_csv("../data/Superstore_Cleaned.csv")
df.head()
```

## Step 3: Display Dataset Information

Display the structure and basic information about the dataset.

```python
df.info()
```

## Step 4: Create Output Folder

Create a folder to store the generated visualizations.

```python
os.makedirs("../visualizations", exist_ok=True)
```

## Step 5: Line Plot

Create a line plot to visualize the total sales trend by category.

```python
sales_by_category = df.groupby("Category")["Sales"].sum()

plt.figure(figsize=(8,5))
plt.plot(sales_by_category.index, sales_by_category.values, marker='o')
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.grid(True)

plt.savefig("../visualizations/line_plot.png")
plt.show()
```

### Observation

The line plot shows the total sales across different product categories, helping identify the categories with higher sales.

## Step 6: Bar Chart

Create a bar chart to compare total sales by category.

```python
sales_by_category = df.groupby("Category")["Sales"].sum()

plt.figure(figsize=(8,5))
plt.bar(sales_by_category.index, sales_by_category.values)
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")

plt.savefig("../visualizations/bar_chart.png")
plt.show()
```

### Observation

The bar chart compares total sales across all product categories, making it easy to identify the highest-performing category.

## Step 7: Scatter Plot

Create a scatter plot to analyze the relationship between Sales and Profit.

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Sales"], df["Profit"])
plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")

plt.savefig("../visualizations/scatter_plot.png")
plt.show()
```

### Observation

The scatter plot illustrates the relationship between sales and profit, helping identify trends and outliers.

## Step 8: Histogram

Create a histogram to visualize the distribution of sales values.

```python
plt.figure(figsize=(8,5))
plt.hist(df["Sales"], bins=20)
plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")

plt.savefig("../visualizations/histogram.png")
plt.show()
```

## Step 9: Correlation Heatmap

Create a heatmap to visualize the correlation between numerical features in the dataset.

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    df[["Sales","Profit","Quantity","Discount"]].corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.savefig("../visualizations/heatmap.png")

plt.show()
```

### Observation

The heatmap displays the correlation between numerical variables such as Sales, Profit, Quantity, and Discount. It helps identify positive and negative relationships among these features.

## Step 10: Pairplot

Create a pairplot to visualize pairwise relationships between numerical features.

```python
pair_plot = sns.pairplot(
    df[["Sales","Profit","Quantity","Discount"]]
)

pair_plot.figure.savefig("../visualizations/pairplot.png")

plt.show()
```

### Observation

The pairplot provides scatter plots and distributions for multiple numerical variables, making it easier to identify trends, clusters, and correlations.

## Step 11: Boxen Plot

Create a boxen plot to analyze the distribution of sales across different product categories.

```python
plt.figure(figsize=(8,5))

sns.boxenplot(
    x="Category",
    y="Sales",
    data=df
)

plt.title("Sales Distribution by Category")

plt.savefig("../visualizations/boxen_plot.png")

plt.show()
```

### Observation

The boxen plot shows the distribution of sales within each product category, helping identify variability, spread, and potential outliers.
